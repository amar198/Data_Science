```python
record_set = []

cols = ['match_id', 'venue', 'match_date', 'team1', 'team2', 'player_of_match', 'toss_winner', \
        'toss_decision', 'umpire1', 'umpire2', 'winning_team', 'win_type', 'tie_resolution_method']
for filename in sorted(glob.glob(os.path.join(os.getcwd(), '../../EDA/Cricket/t20s', '*.yaml'))):
    with open(filename, 'r') as t20:
        #safe_load recognises only standard YAML tags and cannot construct an arbitory Python object.
        outcome = yaml.safe_load(t20)
        
        print(filename[(filename.rfind('/')+1):filename.rfind('.')])
        
        s = filename[(filename.rfind('/')+1):filename.rfind('.')] + ','
        s = s + str(outcome['info']['dates']).replace(',', '-') + ','
        s = s + outcome['info']['venue'].replace(',', '-') + ','
        s = s + outcome['info']['teams'][0] + ','
        s = s + outcome['info']['teams'][1] + ','
        
        if 'player_of_match' in outcome['info']:
            s = s + outcome['info']['player_of_match'][0] + ','
        else:
            s = s + 'N.A, '
            
        if 'toss' in outcome['info']:
            s = s + outcome['info']['toss']['winner'] + ','
            s = s + outcome['info']['toss']['decision'] + ','
        else:
            s = s + 'N.A, N.A, '
            
        s = s + outcome['info']['umpires'][0] + ','
        s = s + outcome['info']['umpires'][1] + ','
        
        if 'winner' in outcome['info']['outcome']:
            s = s + outcome['info']['outcome']['winner'] + ', N.A'
        else:
            
            if outcome['info']['outcome']['result'] == 'tie':
                if 'bowl_out' in outcome['info']['outcome']:
                    s = s + outcome['info']['outcome']['bowl_out']
                elif 'eliminator' in outcome['info']['outcome']:
                    s = s + outcome['info']['outcome']['eliminator']
            else:
                s = s + 'N.A.'
                
            s = s + outcome['info']['outcome']['result'] + ','
        
        reclset = [field for field in s.split(',')]
        record_set.append(reclset)

print(record_set)
```

```python
df = pd.DataFrame(record_set, columns=cols)
df = df.set_index('match_id')
df
```
